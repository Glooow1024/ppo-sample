```python
# 查看保存的 expert 的信息

import h5py
import numpy as np

count = 0
h5_filename = './data/HalfCheetah-v2/trajs_'+'HalfCheetah-v2'+'_%05d.h5'%(count)
#h5_filename = './gail_experts/trajs_halfcheetah.h5'
data = {}

with h5py.File(h5_filename, 'r') as f:
    data['states'] = f['states']
    data['actions'] = f['actions']
    data['rewards'] = f['rewards']
    data['done'] = f['done']
    data['lengths'] = f['lengths']
    print(data['states'].shape)
    print(data['actions'].shape)
    print(data['rewards'].shape)
    print(data['done'].shape)
    print(data['lengths'].shape)
    
    i,j = 90,100
    print('-'*10+' value '+'-'*10)
    print(data['states'][i][j])
    print(data['actions'][i][j])
    print(data['rewards'][i][j])
    print(data['done'][i][j])
```
